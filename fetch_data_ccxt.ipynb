```python
import ccxt
import pandas as pd

# All tickers must be the format ---> 'BASE/QUOTE' <----
# You can add non USDT pairs, such as 'BTC/BNB' etc
# However, that may get confusing.

# get all USDT tickers from binance futures

tickers = list()
for ticker in ccxt.binance().fetch_tickers():
    tickers.append(ticker) if ticker.endswith('USDT') else None

# remove all UP/DOWN BEAR/BULL 
remove_strings = ['UP', 'DOWN', 'BEAR', 'BULL']
for s in remove_strings:
    tickers = [t for t in tickers if s not in t]
tickers.append('SUPER/USDT')


    # Fetch OHLCV data 
    #
    #  @staticmethod
def fetchOHLCVData(symbol: str, timeframe: str, days: int = 30) -> list:
        exchange = ccxt.binance({
           'timeout': 10000,
           'enableRateLimit': True,
           'rateLimit': 250,
           'options': {
               'defaultType': 'future',
            
           }
       })
        exchange.loadMarkets()

        # Get all OHLCV data since
        since = exchange.milliseconds() - days * 24 * 60 * 60 * 1000

        # Data container
        ohlcv_data = []

        # Paginate through data
        while since < exchange.milliseconds():

            #Fetch OHLCV data
            ohlcv =  exchange.fetchOHLCV(symbol, timeframe=timeframe, since=since)

            # Did we received enough valid data
            if ohlcv and len(ohlcv) > 1:
                since = ohlcv[-1][0] + (ohlcv[-1][0] - ohlcv[-2][0])

                # Append OHLCV data
                ohlcv_data += ohlcv
            else:
                break

        # Return OHLCV data
        return ohlcv_data

def data_to_df(data):
    df = pd.DataFrame(data, columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
    df['Timestamp'] = pd.to_datetime(df.Timestamp, unit='ms')
    df = df.set_index('Timestamp')
    return df

# Fetch OHLCV data
# to use with asyncio
# ohlcv = await self._exchange.fetchOHLCV(market.symbol, timeframe=timeframe, since=since)
```
