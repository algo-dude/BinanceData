```python
import pandas as pd
import numpy as np
from functions import *
#%matplotlib inline
filepath = 'fullhistory.csv'


df = pd.read_csv(filepath)

# simplify names
df = df.rename(columns={'Date(UTC)':'Date', 'Pair':'Base', 'Realized Profit':'RealizedProfit'})

# can't do math on strings
df['Fee'] = df['Fee'].str.replace(' USDT', '')
df['Amount'] = df['Amount'].str.replace(' USDT', '')
df['RealizedProfit'] = df['RealizedProfit'].str.replace(',','')
df['Amount'] = df['Amount'].str.replace(',','')
df['Base'] = df['Base'].str.replace('USDT','')

# convert string numbers to float
df = df.astype({'RealizedProfit': float, 'Fee': float, 'Amount': float})

# immediate if - condition, true, false
df['TradeType'] = np.where(df['RealizedProfit'] == 0, "entry", np.where(df['RealizedProfit'] > 0, 'profit', 'loss'))

df.Date = pd.to_datetime(df.Date)
pairs = df.Base.unique()

df
```
